```python
import openpyxl

# Load the workbook
workbook_path = 'Questions_generated.xlsx'  # Replace with your actual file path
wb = openpyxl.load_workbook(workbook_path)
```

```python
import textstat
def calculate_flesch_reading_ease(text):
    """Calculates the Flesch Reading Ease score for given text."""
    if text:
        flesch_score = textstat.flesch_reading_ease(text)
        return round(flesch_score)
    return None

def calculate_flesch_kincaid_grade(text):
    """Calculates the Flesch-Kincaid Grade Level for given text."""
    if text:
        fk_grade = textstat.flesch_kincaid_grade(text)
        return round(fk_grade)
    return None
```

```python
def get_merged_cell_value(ws, cell):
    # Check if the cell is part of a merged range
    for merged_range in ws.merged_cells.ranges:
        if cell.coordinate in merged_range:
            # Return the value from the top-left cell of the merged range
            return ws.cell(merged_range.min_row, merged_range.min_col).value
    return cell.value
```

```python
import csv

NEW_HEADERS = [
    "Grade level",
    "Reading Ease"
]

# Iterate over all sheets in the workbook
for sheet_name in wb.sheetnames:
    ws = wb[sheet_name]
    print(f"\nSheet: {sheet_name}")
    with open(f"{sheet_name}.csv", "w", newline="", encoding="utf-8") as file:
        writer = csv.writer(file)

         # --- Read original Excel headers (row 1) ---
        original_headers = [
            cell.value
            for cell in ws[1][:ws.max_column]
        ]

        # --- Write combined headers ---
        writer.writerow(original_headers + NEW_HEADERS)
    
        for row in ws.iter_rows(min_row=2, max_col=ws.max_column, max_row = 1000):

            row_values = []

            for i, cell in enumerate(row):
                if i == 1:  # Column B (merged cells)
                    row_values.append(
                        get_merged_cell_value(ws, cell)
                    )
                else:
                    row_values.append(cell.value)

            question = row[3].value

            row_values.extend([
                calculate_flesch_kincaid_grade(question),
                calculate_flesch_reading_ease(question),
            ])

            writer.writerow(row_values)
```
